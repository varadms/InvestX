```python

# simple moving average
def sma(period, series):
    sma_points = []
    for i in range(0, len(series) - period + 1):
        sum_value = series[i:i+period].sum()
        mean = sum_value / period
        sma_points.append(mean)
    return sma_points

# exponential moving Average
def ema(period, series):
    sf = 2 / (period + 1)
    ema_points = []
    sma_initial = sma(period, series)[0]
    ema_points.append(series[0] * sf + sma_initial * (1 - sf))
    ema_temp = ema_points[0]
    for i in range(1, len(series)):
        ema_val = series[i] * sf + ema_temp * (1 - sf)
        ema_temp = ema_val
        ema_points.append(ema_val)
    return ema_points

# Relative Strength Index ( momentum oscillator )
def rsi(period, series):
    def gain_loss(series):
        change_list = np.diff(series) # diff between consecutive element 
        gains = np.where(change_list > 0, change_list, 0) # last zero for other elements
        losses = np.where(change_list < 0, -change_list, 0)
        return gains, losses

    def avg_gain_loss(period, series):
        avg = np.mean(series[:period]) # mean of first 'period' element 
        average_gains = [avg] # list and added intial avg
        for i in range(period, len(series)):
            avg_gain_val = ((average_gains[-1] * (period - 1)) + series[i]) / period
            average_gains.append(avg_gain_val)
        return average_gains

    gains, losses = gain_loss(series)
    avg_gains = avg_gain_loss(period, gains)
    avg_losses = avg_gain_loss(period, losses)
    rs = np.array(avg_gains) / np.array(avg_losses)
    rsi = 100 - (100 / (1 + rs))
    return pd.Series(rsi, index=series.index[period:]) # convert np to pandas series , index takes care that index para aligns with original series 

# Moving Average Convergence Divergence
def MACD(series, short_period=12, long_period=26, signal_period=9):
    ema_short = ema(short_period, series) # ema for short period 
    ema_long = ema(long_period, series) # ema for long period 
    macd_line = np.array(ema_short) - np.array(ema_long)
    signal_line = ema(signal_period, macd_line) 
    return pd.DataFrame({'MACD': macd_line, 'Signal': signal_line}, index=series.index)

#Average True Range
def CalcAverageTrueRange(data, period):
    def CalcTrueRange(high, low, cpy):
        a = high - low
        b = abs(high - cpy)
        c = abs(low - cpy)
        return max(a, b, c)
    
    atr_list = []
    for i in range(period, len(data)):
        tr = CalcTrueRange(data['High'].iloc[i], data['Low'].iloc[i], data['Close'].iloc[i-1])
        if i == period:
            atr = tr
        else:
            atr = (atr_list[-1] * (period - 1) + tr) / period
        atr_list.append(atr)
    return pd.Series(atr_list, index=data.index[period:])

# Ichimoku Cloud 
def ic(df, conversion_period=9, base_period=26, leading_span_b_period=52, displacement=26):
    df['ConversionLine'] = (df['High'].rolling(window=conversion_period).max() + df['Low'].rolling(window=conversion_period).min()) / 2
    df['BaseLine'] = (df['High'].rolling(window=base_period).max() + df['Low'].rolling(window=base_period).min()) / 2
    df['LeadingSpanA'] = ((df['ConversionLine'] + df['BaseLine']) / 2).shift(displacement)
    df['LeadingSpanB'] = ((df['High'].rolling(window=leading_span_b_period).max() + df['Low'].rolling(window=leading_span_b_period).min()) / 2).shift(displacement)
    df['LaggingSpan'] = df['Close'].shift(-displacement)
    return df[['ConversionLine', 'BaseLine', 'LeadingSpanA', 'LeadingSpanB', 'LaggingSpan']]

# Ease of Movements
def eom(df, period=14):
    hld = df['High'] - df['Low'] # high low diff
    hla = (df['High'] + df['Low']) / 2 # average
    volume = df['Volume']
    dm = (hla.diff(1) / hld) - (hla.shift(1).diff(1) / hld)
    emv = dm * (volume / (hld * 1000000)) # ease of movement value 
    emv_smoothed = emv.rolling(window=period).mean()
    return pd.Series(emv_smoothed, index=df.index)

# Keltner Channels 
def kc(df, period=20, multiplier=2):
    typical_price = (df['High'] + df['Low'] + df['Close']) / 3
    middle = np.array(ema(period, typical_price))
    atr = np.array(CalcAverageTrueRange(df, period))
    upper = middle[period:] + (multiplier * atr)
    lower = middle[period:] - (multiplier * atr)
    df['Upper'] = pd.Series(upper, index=df.index[period:])
    df['Middle'] = pd.Series(middle, index=df.index)
    df['Lower'] = pd.Series(lower, index=df.index[period:])
    return df[['Upper', 'Middle', 'Lower']]

# Optimize lookback period for Rate of Change(ROC)
best_correlation = -1
best_lookback = -1

for lookback in range(7, 31):
    df['ROC'] = ((df['Close'] - df['Close'].shift(lookback)) / df['Close'].shift(lookback)) * 100
    cleaned_data = df.dropna()
    if len(cleaned_data) > lookback:
        correlation, _ = pearsonr(cleaned_data['ROC'], cleaned_data['Close'])
        if correlation > best_correlation:
            best_correlation = correlation
            best_lookback = lookback

# Calculating ROC with optimized values
df['ROC'] = ((df['Close'] - df['Close'].shift(best_lookback)) / df['Close'].shift(best_lookback)) * 100
```

```python
# Apply the indicators
ic(df)
kc(df)

# Create the dictionary of indicators
dictionary = {
    'rsi': rsi(14, df['Close']),
    'macd': MACD(df['Close']),
    'eom': eom(df),
    'atr': CalcAverageTrueRange(df, 14)
}

# Add indicators to the DataFrame
df['RSI'] = rsi(14, df['Close'])
macd_df = MACD(df['Close'])
df['MACD'] = macd_df['MACD']
df['MACD-SIGNAL'] = macd_df['Signal']
df['EOM'] = eom(df)
df['ATR'] = CalcAverageTrueRange(df, 14)

# Drop NaN values and 'Adj Close' column
df = df.dropna()
df = df.drop(['Adj Close'], axis=1)
```

```python

df['CONV-BASE'] = df['ConversionLine']-df['BaseLine']
df['CONV-BASE'] = df['ConversionLine']-df['BaseLine']
df['a-b']=df['LeadingSpanA']-df['LeadingSpanB']
df['Upper New']=df['Upper']-df['Close']
df['Lower New']=df['Lower']-df['Close']
df['EMA']=ema(4,df['Close'])
df=df.drop(['ConversionLine','BaseLine','Open','LeadingSpanA','LeadingSpanB','Lower','Middle','ATR','Upper' , 'EOM' , 'Upper New' , 'Lower New' ],axis=1)
```

```python
scaler = MinMaxScaler()
scaler = scaler.fit(df)
data_scaled = scaler.transform(df)
train_size=int(len(data_scaled)*0.9)
test_size = len(data_scaled)-train_size
train_scaled,test_scaled = data_scaled[:train_size],data_scaled[train_size:len(data_scaled)]
train_data,test_data = df[:train_size],df[train_size:len(df)]

```

```python
# we will predict with the previous 2 weeks indications 
def to_sequences(df_for_training_scaled,df_for_training,n_past,n_future=1):
    trainx=[]
    trainy=[]
    for i in range(n_past, len(df_for_training_scaled) - n_future +1):
        trainx.append(df_for_training_scaled[i - n_past:i, 1:df_for_training.shape[1]])
        trainy.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

    trainx, trainy = np.array(trainx), np.array(trainy)
    return trainx,trainy

print('trainX shape == {}.'.format(to_sequences(train_scaled,train_data,14)[0].shape))
print('trainY shape == {}.'.format(to_sequences(train_scaled,train_data,14)[1].shape))

```

```python
trainX,trainY = to_sequences(train_scaled,train_data,n_past)
testX,testY = to_sequences(test_scaled,test_data,n_past)

```

```python
# Changing the shape of the dataset as required by LSTM
trainX = trainX.reshape(trainX.shape[0], trainX.shape[1], trainX.shape[2], 1)
trainY = trainY.reshape(trainY.shape[0], 1)
testX=testX.reshape(testX.shape[0], testX.shape[1], testX.shape[2], 1)
testY=testY.reshape(testY.shape[0],1)
trainX.shape
```

* filters=64: The number of convolutional filters (kernels) to apply. Each filter will learn different features from the input data.

* kernel_size=(1,1): The size of the convolutional kernel. A size of (1,1) means the convolution operation will only look at single time steps and single spatial locations, effectively working on a "per frame" basis.

* activation='tanh': The activation function used in this layer. tanh introduces non-linearity, helping the model learn complex patterns.

* input_shape=(trainX.shape[1], trainX.shape[2], 1, 1): Specifies the shape of the input data. Here, it indicates that each input sample has dimensions (n_past, num_features, 1, 1), where 1 is the number of channels. The input_shape is defined by the shape of trainX after reshaping.





```python
# ConvolutionalLSTM
model = tf.keras.Sequential([tf.keras.layers.ConvLSTM2D(filters=64,kernel_size=(1,1),activation='tanh',input_shape=(trainX.shape[1],trainX.shape[2],1,1)),
                            tf.keras.layers.Flatten(),
                            tf.keras.layers.Dense(1000),
                            tf.keras.layers.Dense(500),
                            tf.keras.layers.Dense(100),
                            tf.keras.layers.Dense(1)])
model.compile(optimizer='adam',loss='mean_squared_error')
model.summary()
```

```python

history = model.fit(trainX, trainY, epochs=3,batch_size=2, validation_split=0.2, verbose=1)
```

```python
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
```

```python
predictY = model.predict(testX)
plt.plot(predictY)
plt.plot(testY)
plt.legend(loc='best',labels=['Predicted','True'])
```

```python

def directional_accuracy(y_true, y_pred):
    y_true_direction = np.sign(np.diff(y_true))
    y_pred_direction = np.sign(np.diff(y_pred))
    da = np.mean(y_true_direction == y_pred_direction)
    return da
def calculate_mae(y_true, y_pred):
    mae = np.mean(np.abs(y_true - y_pred))
    return mae

def calculate_rmse(y_true, y_pred):
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    return rmse

def calculate_mape(y_true, y_pred):
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return mape
yo = testY.flatten()
oy = predictY.flatten()
print(f'directional_accuracy = {directional_accuracy(yo,oy)*100}%')
print(f'mae = {calculate_mae(yo,oy)}')
print(f'rmse = {calculate_rmse(yo,oy)}')
print(f'mape = {calculate_mape(yo,oy)}%')

```
