```python
####### Here I have calculated the Volume Price Indicator ###########

import numpy as np
import pandas as pd
import yfinance as yf
import math
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2017-01-01", end="2017-04-30")
```

data['Close'].pct_change(): This calculates the percentage change in the closing price from one period to the next. The pct_change() function computes the difference between the current closing price and the previous closing price, divided by the previous closing price.

```python
data['Price Change'] = data['Close'].pct_change()
data['Price Change'].fillna(0, inplace=True)
```

Volume Price Trend Calculation:

Multiply the volume by the percentage price change to get the volume-adjusted price change for each period.
Compute the cumulative sum of these volume-adjusted price changes to get the VPT.

```python
# Calculate Volume Price Trend
data['VPT'] = (data['Volume'] * data['Price Change']).cumsum()
data
```
